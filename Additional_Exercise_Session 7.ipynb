```python
#Import iris dataset
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
print(iris.head())
```

1)Use  distplot() to see the distribution of the SapelLengthCm,SapelWidthCm,PetalLengthCm,PetalWidthCm . Plot them as subplots in a single image

```python
import seaborn as sb
import matplotlib.pyplot as plt

#Setup the figure
fig, ax = plt.subplots(figsize=(10,10), ncols=2, nrows=2)
#plot for sepal length
sb.distplot(data['sepal length (cm)'],ax=ax[0,0])
#plot for sepal width
sb.distplot(data['sepal width (cm)'],ax=ax[0,1])
#plot for petal length
sb.distplot(data['petal length (cm)'],ax=ax[1,0])
#plot for petal width
sb.distplot(data['petal width (cm)'],ax=ax[1,1])
    
```

2) Do a box plot of all the featres except 'Species'

```python
#Create figure
fig = plt.figure(figsize=(20,7))
#Do boxplot
sb.boxplot(data=data)
```


3 ) Do a count plot for the feature species

```python
#Create figure
fig, ax = plt.subplots(figsize=(7, 3))
#Plot the figure
sb.countplot(y=iris['Species'], data=iris, color="b");
```

4 ) Do a pair plot on features sepal length (cm)  sepal width (cm)  petal length (cm)  petal width (cm),Species

```python
## Pair plot wrt Species:

sb.pairplot(iris,hue='Species',size=3)
```

5) Do an Implot of the following SepalLengthCm ,PetalLengthCm .Using hue ,display the different species in different colour

```python
sb.lmplot(x="sepal length (cm)", y="petal length (cm)", hue="Species", data=iris)
```

6)  Do a bar plot of Species vs SepalLengthcm

```python
sb.barplot(x='sepal length (cm)',y='Species' , data=iris )
```

7) using heatmap , plot correlation matrix

```python
## Correlation plot:
sb.heatmap(iris.corr(),cmap='coolwarm')
plt.title('Iris correlation')
```
